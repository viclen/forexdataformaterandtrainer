```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("rates-p/completehour.csv", index_col="Time")
```

```python
data.index = data.index.to_datetime()
```

```python
period = 200
plt.plot(data.index[-period:], data.Value[-period:])
plt.plot(data.index[-period:], data.SMA[-period:])
plt.plot(data.index[-period:], data.LBB[-period:])
plt.plot(data.index[-period:], data.HBB[-period:])

plt.legend()
```
